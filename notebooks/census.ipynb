```python
import requests

# Fetch variables metadata
url = "https://api.census.gov/data/2019/acs/acs1/pums/variables.json"
response = requests.get(url)
data = response.json()

# This dict has variable names as keys
variables = data["variables"]

# Check if variables exist
test_vars = ["AGS", "PWGTP", "XYZ999"]
for var in test_vars:
    if var in variables:
        print(f"{var}: Found — {variables[var].get('label')}")
    else:
        print(f"{var}: Not found")
```
